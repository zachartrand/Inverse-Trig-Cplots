# Color wheel graphs of inverse trigonometric functions.
These graphs were created with the 
[cplot](https://pypi.org/project/cplot/) module. These were created to 
replace the color wheel graphs made with the hsv color map with graphs 
using a [perceptually uniform](https://colorcet.com/) color map which 
avoids misleading patterns caused by irregular luminance in the color 
map and only shows patterns that are created from the numbers.

The maps are being used on this page: [Inverse Trigonometric Functions](https://en.wikipedia.org/wiki/Inverse_trigonometric_functions)

```python
import numpy as np
import matplotlib.pyplot as plt
import cplot

pi = np.pi
i = complex(0, 1)
```

```python
WIDTH = 8
TICK_RANGE = 2*WIDTH + 1
# LABEL_RANGE = WIDTH + 1
POINTS = 1024
AXIS_TICKS = [i-WIDTH for i in range(TICK_RANGE)]
# TICK_LABELS = [-2**(i) for i in reversed(range(1, LABEL_RANGE))] + [2**i for i in range(LABEL_RANGE)]
fontsize = 5.0
# scaling = lambda x: x / (x + 1)

metadata_png = dict(
    Title = "Domain Coloring",
    Author = "Zach Chartrand",
)

metadata_svg = dict(
    Title = "Domain Coloring",
    Creator = "Zach Chartrand",
    Keywords = ["Complex", "Imaginary", "Domain", 
                "Coloring", "Cartesian", "Graph"],
)

plt.xticks(AXIS_TICKS, size=fontsize)
plt.yticks(AXIS_TICKS, size=fontsize)
plt = cplot.plot(
    lambda z: z,
    (-WIDTH, WIDTH, POINTS),
    (-WIDTH, WIDTH, POINTS),
    add_axes_labels = False,
    # add_colorbars = False,
    # abs_scaling = scaling,  # how to scale the lightness in domain coloring
    contours_abs = 2.0,
    # contours_arg=(-np.pi / 2, 0, np.pi / 2, np.pi),
    # add_colorbars: bool = False,
    # emphasize_abs_contour_1: bool = True,
    # add_axes_labels: bool = True,
    # saturation_adjustment: float = 1.28,
    # min_contour_length = None,
)
# plt.savefig(f"PNGs/colormap.png", dpi=300, metadata=metadata_png)
# plt.savefig(f"SVGs/colormap.svg", metadata=metadata_svg)

plt.show()
```

```python
plt.clf()
WIDTH = 4.0
POINTS = 1024
AXIS_TICKS = [0.125*WIDTH*i - WIDTH for i in range(int(4*WIDTH+1))]
fontsize = 4.0

functions = dict(
    arcsin = np.arcsin, 
    arccos = np.arccos, 
    arctan = np.arctan,
    arcsec = lambda z: np.arccos(1/z),
    arccsc = lambda z: np.arcsin(1/z),
    arccot = lambda z: 0.5*pi - np.arctan(z),
    # arccot_shift = lambda z: -(np.arctan(z) + 0.5*pi),
)

for f_name, f in functions.items():
    plt.clf()
    
    metadata_png = dict(
        Title = f"Domain Coloring of {f_name.title()}",
        Author = "Zach Chartrand",
    )
    
    metadata_svg = dict(
        Title = f"Domain Coloring of {f_name.title()}",
        Creator = "Zach Chartrand",
        Keywords = ["Trigonometry", "Trigonometric", "Inverse", 
                    "Function", "Complex", "Imaginary", "Domain", 
                    "Coloring", "Cartesian", "Graph"],
    )
    plt = cplot.plot(
        lambda z: f(z),
        (-WIDTH, WIDTH, POINTS),
        (-WIDTH, WIDTH, POINTS),
        add_colorbars = False,
        add_axes_labels = False,
        # abs_scaling = scaling,
        # abs_scaling=lambda x: x / (x + 1),  # how to scale the lightness in domain coloring
        contours_abs = 2.0,
        # contours_arg=(-np.pi / 2, 0, np.pi / 2, np.pi),
        # add_colorbars: bool = False,
        # emphasize_abs_contour_1: bool = True,
        # add_axes_labels: bool = True,
        # saturation_adjustment: float = 1.28,
        # min_contour_length = None,
    )
    plt.xticks(AXIS_TICKS, size=fontsize)
    plt.yticks(AXIS_TICKS, size=fontsize)
    # plt.title(f"{f_name}")
    # plt.savefig(f"PNGs/{f_name}.png", dpi=300, metadata=metadata_png)
    # plt.savefig(f"SVGs/{f_name}.svg", dpi=300, metadata=metadata_svg)


    plt.show()
```
